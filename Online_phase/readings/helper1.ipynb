```python
lables=['32']
sets=[32]
# sets=[2]
[]
paths=['sota_single_cpu','sota_all_cpu','random','greedy','greedy_wc','naive_cpu','static_clustering']
# paths=['static_clustering']
thresh_vgg=2.0742
thresh_squeeznet=0.8143
thresh_pilotnet=0.5526
```

```python
globle_list=[]
thresh=2.2202
print('-----------VGG16---------------')
for path in paths:

    print(f'_________________{path}________________')
    for k,i in enumerate(sets):
        no_of_inf=i
        mins=[]
        maxs=[]
        aves=[]
        deadline=[]
        for j in [0,4,8,16,20,18]:
            file=pd.read_csv(f'./{path}/inst_set_{i}_delay/inf{j}_greedy.csv', header=None)
            averages = file.iloc[:, :5].mean()
            minimums= file.iloc[:, :5].min()
            maximums = file.iloc[:, :5].max()
            aves.append(averages[4])
            mins.append(minimums[4])
            maxs.append(maximums[4])
            
            
        t1=np.mean(aves)
        t2=min(mins)
        t3=np.mean(maxs)
        print(f'Readings for INF Set {lables[k]}')
        temp=[t1,t2,t3]
        print(temp)
        globle_list.append(temp)
```
