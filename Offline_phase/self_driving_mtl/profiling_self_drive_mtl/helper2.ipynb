```python
NO_OF_CPU=24
NO_OF_LAYER=18
```

```python
layers_ex=np.genfromtxt('./readings/single_layer_profile_self_driving_mtl_for_percentage.csv', delimiter=',', skip_header=0)
```

```python
conv2conv = np.genfromtxt('./readings/self_driving_mtl_profiling_percentage/conv2conv.csv', delimiter=',', skip_header=0)
conv2flat=np.genfromtxt('./readings/self_driving_mtl_profiling_percentage/conv2flat.csv', delimiter=',', skip_header=0)
fc2act=np.genfromtxt('./readings/self_driving_mtl_profiling_percentage/fc2act.csv', delimiter=',', skip_header=0)
fc2fc=np.genfromtxt('./readings/self_driving_mtl_profiling_percentage/fc2fc.csv', delimiter=',', skip_header=0)
flat2fc = np.genfromtxt('./readings/self_driving_mtl_profiling_percentage/flat2fc.csv', delimiter=',', skip_header=0)
```

```python
print((conv2conv[0]))
print(layers_ex[0])
```

```python
def give_core_ids(layer_desc):
    sorted_indices = np.argsort(layer_desc)

    # Create a new array containing the original indices sorted by the values in the original array
    sorted_original_indices_list = list(sorted_indices)
    
    return sorted_original_indices_list
    
```

```python
NO_OF_CPU=24
NO_OF_LAYERS=18

all_conn=[]
len(all_conn)
```

```python
layer_o=give_core_ids(layers_ex[0])
layer_1=give_core_ids(layers_ex[1])
zero_layer=[]
for i in range(NO_OF_CPU):
    zero_layer.append(layer_o)
    
zero_layer

first_layer=[]
for i in range(NO_OF_CPU):
    first_layer.append(layer_1)
len(layers_ex)
# first_layer
```

```python
global_list=[]
global_list.append(zero_layer)
global_list.append(first_layer)


for i in range(2,NO_OF_LAYERS):
    # percents=None
    if i in [1,2,3,4,5]:
        percents=conv2conv
        print('conv2conv')
    elif i in [6]:
        percents=conv2flat
        print('conv2flat')
    elif i in [14,15,16,17]:
        percents=fc2act
        print('fc2act')
    elif i in [8,9,10,11,12,13]:
        percents=fc2fc
        print('fc2fc')
    elif i in [7]:
        percents= flat2fc
        print('flat2fc')
    

    cur_lay=layers_ex[i]

    computed=[[0]*NO_OF_CPU]*NO_OF_CPU
    for k in range(NO_OF_CPU):
        for j in range(NO_OF_CPU):
            computed[k][j]=cur_lay[j] + (percents[k][j]*cur_lay[j])/100
         
        # print(computed[k])   
        computed[k]=give_core_ids(computed[k])
        
          
    print('===========================')
    global_list.append(computed)


```
