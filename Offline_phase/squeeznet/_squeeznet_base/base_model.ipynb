```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python

def fire_module(x, squeeze_filters, expand_filters):
    squeeze = layers.Conv2D(squeeze_filters, (1, 1), activation='relu', padding='same')(x)
    expand1x1 = layers.Conv2D(expand_filters, (1, 1), activation='relu', padding='same')(squeeze)
    expand3x3 = layers.Conv2D(expand_filters, (3, 3), activation='relu', padding='same')(squeeze)
    return layers.Concatenate()([expand1x1, expand3x3])

def SqueezeNet(input_shape=(224, 224, 3), num_classes=1000):
    inputs = layers.Input(shape=input_shape)

    conv1 = layers.Conv2D(96, (7, 7), strides=(2, 2), activation='relu', padding='same')(inputs)
    maxpool1 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same')(conv1)

    fire2 = fire_module(maxpool1, squeeze_filters=16, expand_filters=64)
    fire3 = fire_module(fire2, squeeze_filters=16, expand_filters=64)
    fire4 = fire_module(fire3, squeeze_filters=32, expand_filters=128)
    maxpool4 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same')(fire4)

    fire5 = fire_module(maxpool4, squeeze_filters=32, expand_filters=128)
    fire6 = fire_module(fire5, squeeze_filters=48, expand_filters=192)
    fire7 = fire_module(fire6, squeeze_filters=48, expand_filters=192)
    fire8 = fire_module(fire7, squeeze_filters=64, expand_filters=256)
    maxpool8 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same')(fire8)

    fire9 = fire_module(maxpool8, squeeze_filters=64, expand_filters=256)
    conv10 = layers.Conv2D(num_classes, (1, 1), activation='relu', padding='same')(fire9)
    global_avgpool10 = layers.GlobalAveragePooling2D()(conv10)
    outputs = layers.Activation('softmax')(global_avgpool10)

    model = models.Model(inputs, outputs)
    return model
```

```python

# Create SqueezeNet model
model = SqueezeNet()
tf.keras.utils.plot_model(model, to_file='squeeznet_model.png', show_shapes=True)
    
#print No of layers ion the model
print(len(model.layers))
# Print model summary
model.summary()
```

```python

def fire_module(x, squeeze_filters, expand_filters,skip_connector=None):
    squeeze = layers.Conv2D(squeeze_filters, (1, 1), activation='relu', padding='same')(x)
    expand1x1 = layers.Conv2D(expand_filters, (1, 1), activation='relu', padding='same')(squeeze)
    expand3x3 = layers.Conv2D(expand_filters, (3, 3), activation='relu', padding='same')(squeeze)
    if skip_connector is not None:
        return layers.Concatenate()([expand1x1, expand3x3, skip_connector])  # Add skip connection
    else:
        return layers.Concatenate()([expand1x1, expand3x3])
def layer_concatenator(x,y):
    return layers.Concatenate()([x, y])
    
def SqueezeNet_1_skip(input_shape=(224, 224, 3), num_classes=1000):
    inputs = layers.Input(shape=input_shape)

    conv1 = layers.Conv2D(64, (3, 3), strides=(2, 2), activation='relu', padding='valid')(inputs)
    maxpool1 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='valid')(conv1)

    fire2 = fire_module(maxpool1, squeeze_filters=16, expand_filters=64)
    fire3 = fire_module(fire2, squeeze_filters=16, expand_filters=64)
    maxpool3 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='valid')(fire3)

    fire4 = fire_module(maxpool3, squeeze_filters=32, expand_filters=128,skip_connector=fire2)
    fire5 = fire_module(fire4, squeeze_filters=32, expand_filters=128)
    maxpool5 = layers.MaxPooling2D((3, 3), strides=(2, 2), padding='valid')(fire5)

    fire6 = fire_module(maxpool5, squeeze_filters=48, expand_filters=192,skip_connector=fire4)
    fire7 = fire_module(fire6, squeeze_filters=48, expand_filters=192)
    fire8 = fire_module(fire7, squeeze_filters=64, expand_filters=256,skip_connector=fire6)
    fire9 = fire_module(fire8, squeeze_filters=64, expand_filters=256)

    conv10 = layers.Conv2D(num_classes, (1, 1), activation='relu', padding='valid')(layer_concatenator(fire8,fire9))
    global_avgpool10 = layers.GlobalAveragePooling2D()(conv10)
    outputs = layers.Activation('softmax')(global_avgpool10)

    model = models.Model(inputs, outputs)
    return model
```

```python

# Create SqueezeNet 1.1 model with skip connections
model_1_1_skip = SqueezeNet_1_skip()
tf.keras.utils.plot_model(model_1_1_skip, to_file='squeeznet_11_skip_model.png', show_shapes=True)
    
#print No of layers ion the model
print(len(model12.layers))
# Print model summary
model_1_1_skip.summary()
```
