```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
NO_OF_CPU=24
NO_OF_LAYER=11
```

```python
layers_ex=np.genfromtxt('./readings/single_layer_profile_squeeznet_for_percentage.csv', delimiter=',', skip_header=0)
```

```python

def make_pecentages(data,fl,out_file_name):
    ave_communication=[[0]*NO_OF_CPU]*NO_OF_CPU
    for i in range(NO_OF_CPU):
        
        for j in range(NO_OF_CPU):
            ave_communication[i][j]=min(layers_ex[fl[0]])+min(layers_ex[fl[1]])
            # ave_communication[i][j]=layers_ex[3][i]+layers_ex[4][j]
        
    for k in range(NO_OF_CPU):
        for jk in range(NO_OF_CPU):
            data[k][jk]=((data[k][jk]-ave_communication[k][jk])/data[k][jk])*100
        # print(data[k])
    # import numpy as np

    def save_2d_list_to_csv(data, file_name):

        with open(file_name, 'w', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerows(data)

        print(f'CSV file "{file_name}" has been created.')
        
    pth=f'./readings/squeeznet_profiling_percent/{out_file_name}.csv'
    save_2d_list_to_csv(data, pth)


    def make_heatmap(readings,name):
        plt.imshow(readings, cmap='viridis', interpolation='nearest')    
        path='./'+name+'.png'
        plt.colorbar()
        plt.title(name)
        plt.savefig(path)
        plt.show()
        
    pths=f'./readings/squeeznet_profiling_percent/{out_file_name}'
    make_heatmap(data,pths)
```

```python
make_pecentages(add2conv,[40,41],'add2conv')
make_pecentages(concat2conv,[6,7],'concat2conv')
make_pecentages(concat2pool,[15,16],'concat2pool')
make_pecentages(concat2add,[10,11],'concat2add')
make_pecentages(conv2concat,[5,6],'conv2concat')
make_pecentages(conv2conv,[3,4],'conv2conv')
make_pecentages(conv2pool,[1,2],'conv2pool')
make_pecentages(pool2act,[42,43],'pool2act')
make_pecentages(pool2add,[35,40],'pool2add')
make_pecentages(pool2conv,[2,3],'pool2conv')
```
