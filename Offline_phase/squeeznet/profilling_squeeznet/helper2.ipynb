```python
add2conv = np.genfromtxt('./readings/squeeznet_profiling_percent/add2conv.csv', delimiter=',', skip_header=0)
concat2conv=np.genfromtxt('./readings/squeeznet_profiling_percent/concat2conv.csv', delimiter=',', skip_header=0)
concat2pool=np.genfromtxt('./readings/squeeznet_profiling_percent/concat2pool.csv', delimiter=',', skip_header=0)
concat2add=np.genfromtxt('./readings/squeeznet_profiling_percent/concat2add.csv', delimiter=',', skip_header=0)
conv2concat = np.genfromtxt('./readings/squeeznet_profiling_percent/conv2concat.csv', delimiter=',', skip_header=0)
conv2conv = np.genfromtxt('./readings/squeeznet_profiling_percent/conv2conv.csv', delimiter=',', skip_header=0)
conv2pool = np.genfromtxt('./readings/squeeznet_profiling_percent/conv2pool.csv', delimiter=',', skip_header=0)
pool2act = np.genfromtxt('./readings/squeeznet_profiling_percent/pool2act.csv', delimiter=',', skip_header=0)
pool2add = np.genfromtxt('./readings/squeeznet_profiling_percent/pool2add.csv', delimiter=',', skip_header=0)
pool2conv = np.genfromtxt('./readings/squeeznet_profiling_percent/pool2conv.csv', delimiter=',', skip_header=0)
```

```python
print((conv2conv[0]))
print(layers_ex[1])
```

```python
def give_core_ids(layer_desc):
    sorted_indices = np.argsort(layer_desc)

    # Create a new array containing the original indices sorted by the values in the original array
    sorted_original_indices_list = list(sorted_indices)
    
    return sorted_original_indices_list
    
```

```python
NO_OF_CPU=24
NO_OF_LAYERS=44

all_conn=[]
len(all_conn)
```

```python
layer_o=give_core_ids(layers_ex[0])
layer_1=give_core_ids(layers_ex[1])
zero_layer=[]
for i in range(NO_OF_CPU):
    zero_layer.append(layer_o)
    
zero_layer

first_layer=[]
for i in range(NO_OF_CPU):
    first_layer.append(layer_1)
len(layers_ex)
# first_layer
```

```python
global_list=[]
global_list.append(zero_layer)
global_list.append(first_layer)


for i in range(2,NO_OF_LAYERS):
    # percents=None
    if i in [11,21,30,40]:
        percents=add2conv
        print('add2conv')
    elif i in [6,25]:
        percents=concat2conv
        print('concat2conv')
    elif i in [15,34]:
        percents=concat2pool
        print('concat2pool')
    elif i in [10,20,29,39]:
        percents=concat2add
        print('concat2add')
    elif i in [4,5,8,9,13,14,18,19,23,24,27,28,32,33,37,38]:
        percents= conv2concat
        print('conv2concat')
    elif i in [3,7,12,17,22,26,31,36]:
        percents= conv2conv
        print('conv2conv')
    elif i in [1,41]:
        percents=conv2pool
        print('conv2pool')
    elif i in [42]:
        percents= pool2act
        print('pool2act')
    elif i in []:
        percents= pool2add
        print('pool2add')
    elif i in [2,16,35]:
        percents= pool2conv
        print('pool2conv')



    cur_lay=layers_ex[i]

    computed=[[0]*NO_OF_CPU]*NO_OF_CPU
    for k in range(NO_OF_CPU):
        for j in range(NO_OF_CPU):
            computed[k][j]=cur_lay[j] + (percents[k][j]*cur_lay[j])/100
         
        # print(computed[k])   
        computed[k]=give_core_ids(computed[k])
        
          
    print('===========================')
    global_list.append(computed)


```
