```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
NO_OF_CPU=24
NO_OF_LAYER=30
```

```python
layers_ex=np.genfromtxt('./readings/single_layer_profile_custom_mtl_for_percentage.csv', delimiter=',', skip_header=0)
```

```python

def make_pecentages(data,fl,out_file_name):
    ave_communication=[[0]*NO_OF_CPU]*NO_OF_CPU
    for i in range(NO_OF_CPU):
        
        for j in range(NO_OF_CPU):
            ave_communication[i][j]=min(layers_ex[fl[0]])+min(layers_ex[fl[1]])
            # ave_communication[i][j]=layers_ex[3][i]+layers_ex[4][j]
        
    for k in range(NO_OF_CPU):
        for jk in range(NO_OF_CPU):
            data[k][jk]=((data[k][jk]-ave_communication[k][jk])/data[k][jk])*100
        # print(data[k])
    # import numpy as np

    def save_2d_list_to_csv(data, file_name):

        with open(file_name, 'w', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerows(data)

        print(f'CSV file "{file_name}" has been created.')
        
    pth=f'./readings/cutom_mtl_profiling_percentage/{out_file_name}.csv'
    save_2d_list_to_csv(data, pth)


    def make_heatmap(readings,name):
        plt.imshow(readings, cmap='viridis', interpolation='nearest')    
        path='./'+name+'.png'
        plt.colorbar()
        plt.title(name)
        plt.savefig(path)
        plt.show()
        
    pths=f'./readings/cutom_mtl_profiling_percentage/{out_file_name}'
    make_heatmap(data,pths)
```

```python
act2bn = np.genfromtxt('./readings/prof_conn/act2bn.csv', delimiter=',', skip_header=1)
bn2drop=np.genfromtxt('./readings/prof_conn/bn2drop.csv', delimiter=',', skip_header=1)
bn2pool=np.genfromtxt('./readings/prof_conn/bn2pool.csv', delimiter=',', skip_header=1)
conv2act=np.genfromtxt('./readings/prof_conn/conv2act.csv', delimiter=',', skip_header=1)
drop2conv = np.genfromtxt('./readings/prof_conn/drop2conv.csv', delimiter=',', skip_header=1)
drop2fc = np.genfromtxt('./readings/prof_conn/drop2fc.csv', delimiter=',', skip_header=1)
drop2flat = np.genfromtxt('./readings/prof_conn/drop2flat.csv', delimiter=',', skip_header=1)
fc2act = np.genfromtxt('./readings/prof_conn/fc2act.csv', delimiter=',', skip_header=1)
flat2fc = np.genfromtxt('./readings/prof_conn/flat2fc.csv', delimiter=',', skip_header=1)
pool2drop = np.genfromtxt('./readings/prof_conn/pool2drop.csv', delimiter=',', skip_header=1)
```

```python
make_pecentages(act2bn,[2,3],'act2bn')
make_pecentages(bn2drop,[22,24],'bn2drop')
make_pecentages(bn2pool,[3,4],'bn2pool')
make_pecentages(conv2act,[1,2],'conv2act')
make_pecentages(drop2conv,[5,6],'drop2conv')
make_pecentages(drop2fc,[24,26],'drop2fc')
make_pecentages(drop2flat,[15,16],'drop2flat')
make_pecentages(fc2act,[18,20],'fc2act')
make_pecentages(flat2fc,[16,18],'flat2fc')
make_pecentages(pool2drop,[4,5],'pool2drop')
```
