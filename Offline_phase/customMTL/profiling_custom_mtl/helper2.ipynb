```python
layers_ex=np.genfromtxt('./readings/single_layer_profile_custom_mtl_for_percentage.csv', delimiter=',', skip_header=0)
```

```python
act2bn = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/act2bn.csv', delimiter=',', skip_header=0)
bn2drop=np.genfromtxt('./readings/cutom_mtl_profiling_percentage/bn2drop.csv', delimiter=',', skip_header=0)
bn2pool=np.genfromtxt('./readings/cutom_mtl_profiling_percentage/bn2pool.csv', delimiter=',', skip_header=0)
conv2act=np.genfromtxt('./readings/cutom_mtl_profiling_percentage/conv2act.csv', delimiter=',', skip_header=0)
drop2conv = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/drop2conv.csv', delimiter=',', skip_header=0)
drop2fc = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/drop2fc.csv', delimiter=',', skip_header=0)
drop2flat = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/drop2flat.csv', delimiter=',', skip_header=0)
fc2act = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/fc2act.csv', delimiter=',', skip_header=0)
flat2fc = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/flat2fc.csv', delimiter=',', skip_header=0)
pool2drop = np.genfromtxt('./readings/cutom_mtl_profiling_percentage/pool2drop.csv', delimiter=',', skip_header=0)
```

```python
print((act2bn[0]))
print(layers_ex[1])
```

```python
def give_core_ids(layer_desc):
    sorted_indices = np.argsort(layer_desc)

    # Create a new array containing the original indices sorted by the values in the original array
    sorted_original_indices_list = list(sorted_indices)
    
    return sorted_original_indices_list
    
```

```python
NO_OF_CPU=24
NO_OF_LAYERS=30

all_conn=[]
len(all_conn)
```

```python
layer_o=give_core_ids(layers_ex[0])
layer_1=give_core_ids(layers_ex[1])
zero_layer=[]
for i in range(NO_OF_CPU):
    zero_layer.append(layer_o)
    
zero_layer

first_layer=[]
for i in range(NO_OF_CPU):
    first_layer.append(layer_1)
len(layers_ex)
# first_layer
```

```python
global_list=[]
global_list.append(zero_layer)
global_list.append(first_layer)


for i in range(2,NO_OF_LAYERS):
    # percents=None
    if i in [2,7,12,20,21]:
        percents=act2bn
        print('act2bn')
    elif i in [22,23]:
        percents=bn2drop
        print('bn2drop')
    elif i in [3,8,13]:
        percents=bn2pool
        print('bn2pool')
    elif i in [1,6,11]:
        percents=conv2act
        print('conv2act')
    elif i in [5,10]:
        percents= drop2conv
        print('drop2conv')
    elif i in [24,25]:
        percents= drop2fc
        print('drop2fc')
    elif i in [15]:
        percents= drop2flat
        print('drop2flat')
    elif i in [18,19,26,27,28,29]:
        percents= fc2act
        print('fc2act')
    elif i in [16,17]:
        percents= flat2fc
        print('flat2fc')
    elif i in [4,9,14]:
        percents= pool2drop
        print('pool2drop')


    cur_lay=layers_ex[i]

    computed=[[0]*NO_OF_CPU]*NO_OF_CPU
    for k in range(NO_OF_CPU):
        for j in range(NO_OF_CPU):
            computed[k][j]=cur_lay[j] + (percents[k][j]*cur_lay[j])/100
         
        # print(computed[k])   
        computed[k]=give_core_ids(computed[k])
        
          
    print('===========================')
    global_list.append(computed)


```
